```python
How to create random data sets.
Sales data -
import pandas as pd
import numpy as np
np.random.seed(42) 
// n==500
data = {
"Product": np.random.choice(["Laptop", "Mobile", "Tablet", "Headphone", "Camera"], 500), 
                    #categorical
    "Price": np.random.randint(5000, 80000, 500),  # numerical integers
    "Quantity": np.random.randint(1, 20, 500),    # numerical integers
    "Category": np.random.choice(["Electronics", "Accessories"], 500),  # categorical
    "Rating": np.random.uniform(1.0, 5.0, 500).round(1),  # numerical float
}
df = pd.DataFrame(data)
print(df.head())

```

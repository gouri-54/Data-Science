```python
#1.load the  titanic dataset
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Titanic2.csv")
df
```

```python
#2.How many passengers are there in the dataset?
    df.shape
```

```python
#3.Find the number of male and female passengers.
    df['Sex'].value_counts()
```

```python
#4.Calculate the average age of passengers (ignore missing values).
    df['Age'].mean()
```

```python
#5.What percentage of passengers survived?
    (df['Survived'].mean() * 100)
```

```python
#6.Find survival rate by gender.
    df.groupby('Sex')['Survived'].mean() * 100
```

```python
#7.Plot a bar chart showing survival rate by gender.
    (df.groupby('Sex')['Survived'].mean() * 100).plot(kind='bar')
plt.ylabel('Survival Rate (%)')
plt.xlabel('Gender')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
#8.Find survival rate by passenger class (Pclass).
    df.groupby('Pclass')['Survived'].mean() * 100
```

```python
#9.Plot a bar chart showing survival rate by passenger class.
    (df.groupby('Pclass')['Survived'].mean() * 100).plot(kind='bar')
plt.xlabel('Pclass')
plt.ylabel('Survived Rate')
plt.title('Survived Rate by Passenger class')
plt.show()
```

```python
#10.Create a new column Family_Size using SibSp and Parch.
df['Family_Size'] = df['SibSp'] + df['Parch'] + 1
df.head()
//SibSp â†’ number of siblings/spouses aboard

Parch â†’ number of parents/children aboard

+ 1 â†’ counts the passenger themself
```

```python
#11.Plot a histogram of passenger ages.
    plt.hist(df['Age'].dropna(), bins=20)
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title('Histogram of Passenger Ages')
plt.show()
```
